```
{
    "messages":
    [
        {
            "role": "user",
            "content": "<image>Table Recognition:"
        },
        {
            "role": "assistant",
            "content": "<table border=\"1\">...</table>",
        }
    ],
    "images":
    [
        "/mnt/dataset/images/table-01.png"
    ]
}
```
